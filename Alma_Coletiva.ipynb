**Visão Geral:**

O projeto Alma Coletiva - SOS Rio Grande do Sul é uma resposta direta às necessidades urgentes das comunidades atingidas por enchentes. Nosso foco é garantir que informações vitais sobre recursos essenciais estejam ao alcance de todos, quando mais precisam.

**Objetivo Principal:**

Desenvolver um chatbot acessível, servindo como assistente virtual para fornecer informações precisas e atualizadas sobre a disponibilidade de itens fundamentais como alimentos, água potável, medicamentos, roupas, além de abrigos e outros serviços.

**Estratégias de Ação:**

- Centralização de Informações: Uma única plataforma para acessar dados confiáveis sobre recursos e serviços de ajuda.
- Atualizações Dinâmicas: Informações em tempo real para atender às necessidades emergentes das comunidades.
- Integração Comunitária: Conexão efetiva entre afetados, distribuidores, varejistas, pontos de auxílio e autoridades.
- Otimização de Recursos: Melhoria na distribuição de recursos, assegurando apoio rápido e organizado.

**Impacto Esperado:**

A implementação do **Alma Coletiva - SOS Rio Grande do Sul** pretende fortalecer a capacidade de recuperação das comunidades, proporcionando uma resposta coordenada e eficiente em tempos de crise.

## Descrição dos Campos do CSV

1. **Categoria**: Tipo de serviço ou recurso oferecido pela entidade (ex: Supermercado, Farmácia, Centro de Ajuda Humanitária) ** Ver mais sobre Descrição das cotegorias.

2. **NomeEntidade**: Nome oficial da entidade, organização ou estabelecimento.

3. **Endereço**: Endereço completo da entidade, incluindo rua, número, complemento, bairro e municipio. Seguindo o formato "Rua, Número - Bairro, Município". Exemplo: "Av. Assis Brasil, 1234 - Passo D'Areia, Porto Alegre". Esta formatação ajuda na identificação precisa da localização.

4. **Localização Latitude**: Coordenada geográfica da latitude da entidade, facilitando a localização por GPS.

5. **Localização Longitude**: Coordenada geográfica da longitude da entidade, complementando a informação para localização precisa.

6. **Contato**: Números de telefone, e-mail ou outros meios de contato com a entidade.

7. **HorarioFuncionamento**: Dias e horários em que a entidade está aberta e disponível para o público.

8. **ItensDisponiveis**: Lista de itens que podem ser encontrados ou adquiridos na entidade (ex: alimentos, medicamentos, roupas).

9. **ServicosOferecidos**: Descrição dos serviços que a entidade oferece (ex: consultas médicas, distribuição de alimentos, abrigo temporário).

10. **DataAtualizacao**: Data da última atualização das informações fornecidas, indicando a relevância e precisão dos dado.

```python
# Importando Bibliotecar para obter o dado.
import requests
import io
import csv

# URL do arquivo CSV (substitua pelo URL real)
spreadsheet_url = 'https://docs.google.com/spreadsheets/d/e/2PACX-1vQdSmnsM6koFN3XD1nt6dNgtNuUadMwSJClNCcvJ6U5shsaWYSifaYas8KOHqoLg0y-M3kkVhaMQwwx/pub?output=csv'

# Faz o download do arquivo CSV
response = requests.get(spreadsheet_url)

# Verifica se o download foi bem-sucedido
response.raise_for_status()

# Lê o conteúdo do CSV usando o módulo csv
csv_data = csv.reader(io.StringIO(response.text))

# Imprime os dados
for row in csv_data:
  print(row)
```

Ajustando os Caracteres

```python
### PROBLEMAS DE CARACTERES
# Decodifica o texto da resposta usando a codificação 'latin-1' (ajuste se necessário)
decoded_text = response.text.encode('latin-1').decode('utf-8', errors='replace')

# Cria o objeto StringIO com o texto decodificado
csv_data = list(csv.reader(io.StringIO(decoded_text)))

# Imprime os dados
for row in csv_data:
  print(row)
```

Vamos converter em JSON, para user os dataForms da aula 5.

```python
import json

# Extrai os cabeçalhos da primeira linha
headers = csv_data[0]

# Cria uma lista de dicionários
json_data = []
for row in csv_data[1:]:  # Começa da segunda linha
  json_data.append(dict(zip(headers, row)))

# Converte para JSON
json_output = json.dumps(json_data, indent=4)

# Imprime o JSON
print(json_output)
```

```python
# Bibliotecas padão trabalhar com dados.
import pandas               as pd
import numpy                as np

#usado para converter uma string JSON (json_output) de volta para um objeto Python (lista ou dicionário).
json_data = json.loads(json_output)

# **Cria o DataFrame diretamente a partir de json_data (lista de dicionários)**
df = pd.DataFrame(json_data)

# Define os nomes das colunas
df.columns = ["Categoria", "NomeEntidade", "Endereco", "Latitude",
              "Longitude", "Contato", "HorarioFuncionamento",
              "ItensDisponiveis", "ServicosOferecidos", "DataAtualizacao"]

# Agora você pode trabalhar com o DataFrame df
df
```

Mudando Para o modelo gemini-1.5-pro-latest

```python
# Nome do modelo
model_name = "gemini-1.5-pro-latest"

# Define as configurações de segurança
from google.generativeai.types import HarmCategory, HarmBlockThreshold

safety_settings={
        HarmCategory.HARM_CATEGORY_HATE_SPEECH: HarmBlockThreshold.BLOCK_NONE,
        HarmCategory.HARM_CATEGORY_HARASSMENT: HarmBlockThreshold.BLOCK_MEDIUM_AND_ABOVE,
        HarmCategory.HARM_CATEGORY_SEXUALLY_EXPLICIT: HarmBlockThreshold.BLOCK_MEDIUM_AND_ABOVE,
        HarmCategory.HARM_CATEGORY_DANGEROUS_CONTENT: HarmBlockThreshold.BLOCK_MEDIUM_AND_ABOVE
}

generation_config = {
  "temperature": 1,
  "top_p": 0.95,
  "top_k": 0,
  ##"response_mime_type": "application/json", ## Retorno em JSON
}

# Cria um modelo generativo com os atributos especificados
system_instruction = "Gere respostas positivas e seguras."

modelo2 = genai.GenerativeModel(model_name=model_name,
                              generation_config=generation_config,
                              system_instruction=system_instruction,
                              safety_settings=safety_settings)

prompt = f"Reescreva esse texto de uma forma mais amigavel, adicionar o mapa do local com os dados latitude e Longitude: {trecho}"

resposta = model_2.generate_content(prompt)
to_markdown(resposta.text)
```
